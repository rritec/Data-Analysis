### Method 3: by using loc
* Access a group of rows and columns by label(s) or a boolean array

#### Required value Reading

```python
df.loc["Jan","eggs"]
```

#### Slice

* .. warning:: Note that contrary to usual python slices, **both** the
      start and the stop are included

```python
df.loc["Jan":"Mar","eggs"] # slice
```

```python
df.loc["Jan":"Mar","eggs":"spam"] # slice
```

#### Fancy

```python
df.loc[["Jan","Mar"],["eggs","spam"]] # fancy
```

```python
df
```

#### boolean

```python
df.loc[[True,False,False,True,False,True]]
```

```python
df.iloc[[True,False,False,True,False,True]]
```
